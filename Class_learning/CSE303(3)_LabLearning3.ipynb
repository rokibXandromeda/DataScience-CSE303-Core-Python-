```python
list1 = [['Alice',23,3.5,10],['Bob',24,3.4,6],['Charlie',22,3.9,8]]
df = pd.DataFrame(list1)
df
```

```python
df.columns = ['name', 'age', 'cgpa', 'hours']
df
```

```python
df.index = [f'student{i}' for i in range(1, len(list1) + 1)]
df
```

```python
list1 = [['Alice',23,3.5,10],['Bob',24,3.4,6],['Charlie',22,3.9,8]]
columns = ['name', 'age', 'cgpa', 'hours']
indices = [f'student{i}' for i in range(1, len(list1) + 1)]
df1 = pd.DataFrame(data = list1, columns = columns, index = indices)
df1
```
